线性回归代码实战

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#回归，有监督的机器学习，X，y
X =2*np.random.rand(100, 1)
print(len(X))
print(X)
#y是真实的数据 也就是y_hat + error 要模拟出来
y = 5 + 4*X + np.random.randn(100,1)
```

```python
#为了去求解截距项w0，我们给X一开始加上一列全为1的X0
X_b = np.c_[np.ones((100,1)), X]
#np.c_是NumPy中的一个对象，用于连接数组。它可以将两个数组沿着第二个数组的最后一个轴进行连接，类似于按列连接。
#实现解析解的公式来求解θ
θ = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
θ
#不可能是完美解 因为有误差项 有噪音
```

```python
#使用模型去预测
X_new = np.array(
    [[0], [2]]
)
X_new_b = np.c_[(np.ones((2,1))), X_new]
y_predict = X_new_b.dot(θ)
print(y_predict)
```

```python
#绘图
plt.plot(X_new, y_predict, 'r-')
plt.plot(X, y, 'b.')
plt.axis([0, 2, 0 ,15])
plt.show()
```

scikit-learn简洁实现

```python
from sklearn.linear_model import LinearRegression
```

```python
X1 = 2*np.random.rand(100,1)
X2 = 2*np.random.rand(100,1)
X = np.c_[X1, X2]

y = 4 + 3*X1 + 5*X2 +np.random.randn(100,1)
```

```python
reg = LinearRegression()
reg.fit(X, y)
print(reg.intercept_, reg.coef_)
```

```python
X_new = np.array([[0,0],[2,1], [2,4]])
y_predict = reg.predict(X_new)

plt.plot(X_new[:, 0], y_predict, 'r-')
plt.plot(X, y, 'b.')
plt.axis([0, 2, 0 ,25])
plt.show()
```
