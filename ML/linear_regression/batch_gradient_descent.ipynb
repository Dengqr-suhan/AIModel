```python
import numpy as np
```

```python
#创建数据集X， y
X = np.random.rand(100,1)
y =4 + 3*X + np.random.randn(100,1)
X_b = np.c_[np.ones((100,1)), X]
```

```python
#创建超参数
learning_rate  = 0.001
n_iterations = 10000
t0, t1 =5,500
```

```python
#让学习率动态变化
def learning_rate_schedule(t):
    return t0/(t+t1)
```

```python
#1.初始化theta，w0...wn, 正态分布创建w
theta = np.random.randn(2, 1)

#4.判断是否收敛 一般不会去设置阈值 而是采用设置相对较大的迭代次数
for _ in range(n_iterations):
    #2.求梯度,计算gradient
    gradients = X_b.T.dot((X_b.dot(theta)-y))
    #3,应用梯度下降法的公式调整Theta数值 θt + 1 = θt - learning_rate * gradient
    learning_rate = learning_rate_schedule(_)
    print(learning_rate)
    theta = theta - learning_rate * gradients
theta
```
